### 1° vista.

A base de dados possui 51.913 linhas e 8 colunas, onde trazem as reviews do app da amazon na Play Store, com isso, podemos levantar alguns questionamentos.

1.   Quantas reviews postivias e negativas tivemos?
2.   Sobre o que se tratam o review?
3.   Quando foram feitos comentarios?
4.   Sobre qual versão do App eles se tratam?

Porém, antes irei rastrear e analisar alguns possiveis erros e limpar um pouco da planilha, por hora filtrarei a planilha atras de todos os `NaN`.


```python
# Fazendo a tratativa de dados.
# Primeiro, farei uma copia do dataframe para analise eliminaremos os na.

df_amazon_analise = df_amazon.copy()
df_amazon_analise.dropna(inplace=True)
df_amazon_analise.shape
```

```python
# Ficando com 46.868 linhas, abrirei o df.

df_amazon_analise.head()
```

```python
# A paritir de agora, vamos as perguntas.
# 1. Quantas reviews postivias e negativas tivemos?

df_amazon_analise['score'].value_counts()
```

```python
# Tornando esse dado visual

import plotly.express as px

fig = px.pie(df_amazon_analise, values='score', names='score')
fig.show()
```

```python
# Certo, vamos usar as datas para uma melhor visualização

df_amazon_analise.rename(columns={'at': 'data'}, inplace=True)
df_amazon_analise.head()
```

```python
df_amazon_analise.insert(1, 'year', df_amazon_analise['data'].dt.year)
df_amazon_analise.head()
```

```python
balance = df_amazon_analise['score'].value_counts()
balance
```

```python
fig=px.bar(df_amazon_analise, x=balance.index, y=balance.values)
fig.show()
```

```python
fig = px.histogram(df_amazon_analise, x='year', color='score')
fig.show()
```
